# Buildind Model

```python
y = df[['Close']]
x = df[['Open','High','Low','Last','Prev Close']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 , random_state  = 0)
```

```python
#dimentions of training models
x_train.shape
```

```python
#dimentions of testing models
x_test.shape
```

```python
model = LinearRegression()

# Train the Linear Regression Model
model.fit(x_train, y_train)

print('Intercept: ' ,  model.intercept_)
print('coefficient : ' ,  model.coef_)
```

```python
#Predicting the model 
preds = model.predict(x_test)

#print(preds)
#print({'Actual': y_test, 'Predicted': preds})
```

# Predictions

```python
from sklearn.metrics import r2_score

# Evaluate the model's accuracy
accuracy = r2_score(y_test, preds)
print("Accuracy:", accuracy)
```

# Model Evaluation

```python
mean_squared_error(y_test, preds )
```

```python
sns.regplot( x = y_test, y = preds , data = df)
plt.title('Relatioship of Y_test and Yhat')
plt.ylabel('y_test')
plt.xlabel('yhat')
```

```python
# Plot the actual vs. predicted prices for the test set
plt.figure(figsize=(12, 6))
plt.title('Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.plot(df.index[-len(y_test):], y_test, label='Actual Prices')
plt.plot(df.index[-len(y_test):], preds, label='Predicted Prices')
plt.legend()
plt.show()
```
